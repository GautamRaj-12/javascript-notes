- ***XMLHttpRequest***
 - XMLHttpRequest (XHR) objects are used to interact with servers. You can retrieve data from a URL without having to do a full page refresh. This enables a Web page to update just part of a page without disrupting what the user is doing.
 - Despite its name, XMLHttpRequest can be used to retrieve any type of data, not just XML.
 - The XMLHttpRequest (XHR) object in JavaScript is instrumental for interacting with servers, allowing you to fetch data from a URL without requiring a complete page refresh. This functionality empowers web pages to update specific parts of the content seamlessly, enhancing the user experience.
 - To employ XMLHttpRequest, you start by creating an instance of the object using the new keyword and store it in a variable. Afterward, you utilize the open method, providing two parameters: the HTTP method (GET, POST, etc.) and the URL to which the request is directed.
 - However, merely opening the request does not initiate it. You need to invoke the send() method to trigger the actual sending of the request.
 - Once the request is dispatched, the status code undergoes changes as the server processes the request. To monitor these changes dynamically, you can utilize the onreadystatechange event. By assigning a function to this event, you can track the evolving state of the XMLHttpRequest object.
 - Here's a simplified depiction of the different states:



- By keeping an eye on the state changes through onreadystatechange, you can handle the response appropriately, whether it's successful or encounters an error.

 > Pehle XMLHttpRequest ka object banayenge new keyword ka use kar ke aur variable me store kareng
 > Uske baad open function ke andar 2 paramaters denge : method and requested url.
 > Par sirf open karne se request jaati nahi hai call karna padega
 > send() method call karne par wo send hota hai
 > Jab request jaati hai to status code change hoti hai

   


 > humlog is state ko continuously track karenge using `onreadystatechange`
 > iske andar bas ek function de dena hai

```typescript
// Define the URL for the GitHub API endpoint that provides information about a user
const requestUrl = 'https://api.github.com/users/GautamRaj-12';

// Create a new XMLHttpRequest object
const xhr = new XMLHttpRequest();

// Open a new GET request to the specified URL
xhr.open('GET', requestUrl);

// Define a callback function to be executed when the readyState of the XMLHttpRequest changes
xhr.onreadystatechange = function () {
    // Log the current readyState to the console
  console.log(xhr.readyState);
    // Check if the readyState is 4, which means the operation is complete
  if (xhr.readyState === 4) {
      // Parse the responseText property of the XMLHttpRequest as JSON
    const data = JSON.parse(this.responseText);
      // Log the parsed data (GitHub user information) to the console
    console.log(data);
  }
};
// Send the XMLHttpRequest (initiate the request to the specified URL)
xhr.send();
```

## Project : Fetching data from Github API using `XMLHttpRequest()`

```typescript
const requestUrl = 'https://api.github.com/users/GautamRaj-12';

const xhr = new XMLHttpRequest();

xhr.open('GET', requestUrl);

xhr.onreadystatechange = function () {
  console.log(xhr.readyState);
  if (xhr.readyState === 4) {
    const data = JSON.parse(this.responseText);
    console.log(data);
    const { avatar_url, bio, name, login } = data;

    //creating card using this data
    const heading = document.createElement('h1');
    heading.innerHTML = 'Github Card';
    document.body.appendChild(heading);

    const card = document.createElement('div');
    document.body.appendChild(card);

    // card.style.width = '20rem';
    // card.style.textAlign = 'center';
    // card.style.boxShadow = 'rgba(0, 0, 0, 0.35) 0px 5px 15px';
    // card.style.borderRadius = '10px';
    // card.style.padding = '1rem';

    const cardStyles = {
      width: '20rem',
      textAlign: 'center',
      boxShadow: 'rgba(0, 0, 0, 0.35) 0px 5px 15px',
      borderRadius: '10px',
      padding: '1rem',
    };
    Object.assign(card.style, cardStyles);

    const avatar = document.createElement('div');
    avatar.innerHTML = `<img src=${avatar_url} height='100px' width='100px'/>`;
    card.appendChild(avatar);

    const bioData = document.createElement('h3');
    bioData.innerHTML = bio;
    card.appendChild(bioData);

    const displayName = document.createElement('p');
    displayName.innerHTML = `Name: ${name}`;
    card.appendChild(displayName);

    const userName = document.createElement('p');
    userName.innerHTML = `Username: ${login}`;
    card.appendChild(userName);
  }
};
xhr.send();
```
