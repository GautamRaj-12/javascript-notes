## Introduction

- Javascript is an ***asynchronous*** language.
- But it's default nature is ***synchronous*** language.
- It is ***single threaded***.

## Execution Context
- execute one line of code at a time.

## Blocking Code vs Non Blocking Code
- **Blocking Code**
  - Blocks the flow of program. Ex- Read file Sync
- **Non Blocking Code**
  - Does not block execution. Ex - Read file Async
